```python
import pygame
import random

pygame.init()

width = 800
height = 600
clock = pygame.time.Clock()
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption("Shooting Game")
background = pygame.image.load('bg.jfif')
player_image = pygame.image.load('space.png')
enemy_image = pygame.image.load('alien.png')

# Player
spaceshipx = 370
spaceshipy = 480
spaceship_speed = 10
changex = 0
# Enemies
enemies = []
num_enemies = 1
for _ in range(num_enemies):
    enemyx = random.randint(0, 736)
    enemyy = random.randint(30, 150)
    enemies.append([enemyx, enemyy])
enemy_speed = 10
# Bullet
bulletx = 0
bullety = 480
bullet_speed = 20
bullet_state = "ready"
# Score
score = 0
game_over = False

running = True
while running:
    screen.blit(background, (0, 0))

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    if not game_over:
        # Move player
        keys = pygame.key.get_pressed()
        if keys[pygame.K_LEFT]:
            changex = -spaceship_speed
        elif keys[pygame.K_RIGHT]:
            changex = spaceship_speed
        else:
            changex = 0
        spaceshipx += changex
        if spaceshipx <= 0:
            spaceshipx = 0
        elif spaceshipx >= 736:
            spaceshipx = 736
        # Move enemies
        for enemy in enemies:
            enemy[0] += enemy_speed
            if enemy[0] <= 0 or enemy[0] >= 736:
                enemy_speed = -enemy_speed
                enemy[1] += 40
        # Fire bullet
        if keys[pygame.K_SPACE] and bullet_state == "ready":
            bulletx = spaceshipx +16
            bullet_state = "fire"
        if bullet_state == "fire":
            bullety -= bullet_speed
            pygame.draw.rect(screen, (255, 0, 0), (bulletx, bullety, 8, 16))
            #collision
            for enemy in enemies:
                if enemy[0] <= bulletx <= enemy[0] + 64 and enemy[1] <= bullety <= enemy[1] + 64:
                    bullet_state = "ready"
                    bullety = 480
                    score += 1
                    # enemy at a random position
                    enemies.remove(enemy)
                    for _ in range(num_enemies):
                        enemyx = random.randint(0, 736)
                        enemyy = random.randint(30, 150)
                        enemies.append([enemyx, enemyy])
            # Reset bullet position when it reach the top
            if bullety <= 0:
                bullet_state = "ready"
                bullety = 480
        # Draw player and enemies
        screen.blit(player_image, (spaceshipx, spaceshipy))
        for enemy in enemies:
            screen.blit(enemy_image, (enemy[0], enemy[1]))
        # Display
        font = pygame.font.Font(None, 36)
        score_text = font.render("Score: " + str(score), True, (255, 255, 255))
        screen.blit(score_text, (10, 10))
        if any(enemy[1] >= height - 64 for enemy in enemies):
            print("Game Over!")
            game_over = True

    pygame.display.update()
    clock.tick(30)

pygame.quit()
```
